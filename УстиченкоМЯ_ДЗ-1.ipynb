### **Нахождение средних букв**

```python
word = 'text'
```

```python
length = len(word) # длина слова
middle = length // 2 # середина слова, при 4 и 5 буквах выводит 2, при 6 уже 3
```

```python
if length % 2 == 1: # если есть остаток от деления на 2, то слово имеет нечетное количество букв
  print(word[middle]) # вывод средней буквы
else: # если остатка нет, то слово имеет четное количество букв
  print(word[middle-1] + word[middle]) # вывод двух средних букв
```

### **Списки парней и девушек**

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
```

```python
if len(boys) != len(girls): # проверка, равны ли по длине списки
  print("Внимание, кто-то может остаться без пары!")
else:
  boys_sorted = sorted(boys) # новый отсортированный список парней
  girls_sorted = sorted(girls) # новый отсортированный список девушек
```

```python
print("Идеальные пары:")
for i in range(len(boys_sorted)): # вывод отсортированных списков попарно
  print(boys_sorted[i], "и", girls_sorted[i])
```

```python
from typing import List, Tuple

def egcd(a, b):
    if b == 0:
        return 1, 0, a
    x1, y1, g = egcd(b, a % b)
    return y1, x1 - (a // b) * y1, g

def modinv(a, m):
    x, y, g = egcd(a, m)
    if g != 1:
        raise ValueError("Обратный элемент не существует")
    return x % m

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def are_coprime(a, b):
    return gcd(a, b) == 1

def choose_min_e(phi, start=3):
    e = start if start % 2 == 1 else start + 1
    while e < phi:
        if are_coprime(e, phi):
            return e
        e += 2
    raise ValueError("Не найден подходящий e")

def rsa_keygen(p, q):
    n = p * q
    phi = (p - 1) * (q - 1)
    e = choose_min_e(phi, start=3)
    d = modinv(e, phi)
    return n, phi, e, d

def compute_block_size(n):
    """Максимальный размер блока в байтах, чтобы 256^k < n."""
    k = 0
    while 256 ** (k + 1) < n:
        k += 1
    return k

def pack_bytes_to_blocks(data: bytes, block_size: int) -> List[Tuple[int, int]]:
    blocks = []
    for i in range(0, len(data), block_size):
        chunk = data[i:i + block_size]
        val = int.from_bytes(chunk, 'big')
        blocks.append((val, len(chunk)))
    return blocks

def unpack_blocks_to_bytes(blocks: List[Tuple[int, int]]) -> bytes:
    out = bytearray()
    for val, length in blocks:
        chunk = val.to_bytes(length, 'big')
        out.extend(chunk)
    return bytes(out)

def encrypt_message(text: str, e: int, n: int) -> Tuple[List[Tuple[int, int]], int]:
    data = text.encode('utf-8')
    block_size = compute_block_size(n)
    blocks = pack_bytes_to_blocks(data, block_size)
    cipher = [(pow(val, e, n), length) for val, length in blocks]
    return cipher, block_size

def decrypt_message(cipher: List[Tuple[int, int]], d: int, n: int) -> str:
    blocks = [(pow(c, d, n), length) for c, length in cipher]
    data = unpack_blocks_to_bytes(blocks)
    return data.decode('utf-8')


p, q = 233, 293  # вариант №10
n, phi, e, d = rsa_keygen(p, q)

print("=== Ключевые параметры RSA ===")
print(f"p = {p}, q = {q}")
print(f"n = {n}")
print(f"φ(n) = {phi}")
print(f"e = {e}, d = {d}")
print(f"Взаимно просты (e, φ) → {are_coprime(e, phi)}")

print("\nВведите текст для шифрования:")
plaintext = input(">>> ")

if not plaintext.strip():
    print("Пустая строка. Завершение работы.")
    exit(0)

cipher, block_size = encrypt_message(plaintext, e, n)
print("\n=== Результаты шифрования ===")
print(f"Размер блока: {block_size} байт")
print(f"Шифртекст (целые блоки):")
print(" ".join(str(c) for c, _ in cipher))

recovered = decrypt_message(cipher, d, n)
print("\n=== Расшифрование ===")
print("Результат:", recovered)

if recovered == plaintext:
    print("Проверка успешна: тексты совпадают.")
else:
    print("Ошибка: тексты не совпадают!")
```
